# Food Security Analysis for Karamoja, Uganda





## Overview

Karamoja is the most food-insecure region of Uganda due to low crop productivity caused by intense droughts, pest infestations and disease outbreaks. Although NGOs provide farm inputs and technical support, they lack a clear regional overview and often rely on local information to prioritize their activities. To address this gap, our team at DDI was tasked with developing a Food Security Monitoring Tool. Using satellite imagery, the tool estimates yields of the region’s two main staple crops — sorghum and maize — starting with the 2017 season, providing NGOs with better insights for decision-making and prioritization.


## Project Goal

This project will explore the data to achieve the below:

* Providing NGOs with crop yields for Maize and Sorghum across districts and sub-counties.
* Support better allocation of resources and prioritizing interventions.
* Better understanding of food security patterns.

## Data Understanding

This project uses datasets from Dalberg Data Insights measuring the yield of the two main staple crops of the region (Maize and Sorghum)  based on satellite images for the 2017 crop season.

Includes boundaries on Uganda sub-counties and districts, crop type map for sorghum and maize, yield and population per sub-county and district, average yield for sorghum and maize for sub-counties and districts in Hectares, total crop area for sorghum and maize for sub-counties and districts and total productivity for maize and sorghum for sub-counties and districts.

The data will be cleaned and analyzed to extract actionable insights that will help NGOs make data driven decisions.

Creating a new data set with field areas in Hectares per subcounties (Aggregating field area per subcounty). Done by merging subcounty df_1 with sorghum and maize field boundaries df_3 and df_4 (Spatial field joining).

```python
subcounty = df_1.to_crs("EPSG:32636")   #
maize = df_4.to_crs(subcounty.crs)
sorghum = df_3.to_crs(subcounty.crs)
```

```python
maize["Field_Area_Ha"] = maize.geometry.area / 10_000
sorghum["Field_Area_Ha"] = sorghum.geometry.area / 10_000
```

```python
import geopandas as gpd

maize_joined = gpd.sjoin(maize, subcounty, how="inner", predicate="within")
sorghum_joined = gpd.sjoin(sorghum, subcounty, how="inner", predicate="within")
```

```python
maize_summary = maize_joined.groupby("SNAME2014")["Field_Area_Ha"].sum().reset_index()
maize_summary.rename(columns={"Field_Area_Ha": "Maize_Field_Area_Ha"}, inplace=True)

sorghum_summary = sorghum_joined.groupby("SNAME2014")["Field_Area_Ha"].sum().reset_index()
sorghum_summary.rename(columns={"Field_Area_Ha": "Sorghum_Field_Area_Ha"}, inplace=True)
```

```python
subcounty = subcounty.merge(maize_summary, on="SNAME2014", how="left")
subcounty = subcounty.merge(sorghum_summary, on="SNAME2014", how="left")

```

Checking on correlation

```python
plt.figure(figsize=(8,6))
sns.heatmap(subcounty_df[["M_Yield_Ha", "S_Yield_Ha", "M_Area_Ha", "S_Area_Ha", "POP"]].corr(),
            annot=True, cmap="YlGnBu")
plt.title("Correlation Heatmap (Subcounty Level)")
plt.show()
```

#### Observation

Maize and Sorghum yields per hectare (M_Yield_Ha vs S_Yield_Ha) are positively correlated (0.69) → areas performing better in maize also tend to do better in sorghum.

Population (POP) has weak correlations with yields but shows a moderate correlation with sorghum area (S_Area_Ha = 0.43). This means large populations do not automatically translate to higher productivity.

Cultivated area and yields are not strongly correlated, suggesting that expansion of land is not enough; productivity depends on other factors (drought, inputs, soil quality, pests).

# Conclusion

Karamoja is sorghum-dependent—if sorghum fails due to drought, food security will be severely threatened.

Sorghum, traditionally more drought-resistant, is underperforming → NGOs could investigate whether pest/disease outbreaks or poor seed varieties are driving the low yields even though generally it produces more in kg as compared to maize its yield per hectares is wanting. This is an issue that needs to be looked into to avoid subjecting vast lands for sorghum cultivation and neglecting maize.

District-level differences (e.g., Moroto and Abim underperforming) suggest targeted interventions are needed rather than uniform support across all districts.

Low correlation between population and yield highlights the risk of hidden food insecurity hotspots—some densely populated subcounties(e.g subcounties in Kaabong) may be at high risk even if yields look average.

## Recommendations for NGOs


Diversify Beyond Sorghum

* Promote alternative drought-tolerant crops such as maize, cassava, millet, and legumes to reduce dependency on sorghum.

* Support seed distribution programs for these alternatives.

Boost Productivity per Hectare

* Provide training on modern agronomic practices (fertilizer use, pest management, intercropping).

* Introduce improved sorghum varieties with higher yield per hectare and pest resistance.

Target Underperforming Districts (Abim and Moroto)

* Prioritize interventions here by supplying inputs, irrigation tools, and extension services.

* Encourage kitchen gardens and small-scale irrigation to reduce reliance on rain-fed farming.

Address Misallocation of Land

*  Sensitize communities on balancing sorghum cultivation with other crops (e.g., maize, which performs well on smaller plots).

* Support land-use planning and farmer cooperatives to optimize crop mix.

Link Production to Food Security Hotspots

* Establish food banks and grain storage in densely populated but food-insecure districts (like Kaabong).

* Strengthen early-warning systems that connect yield monitoring with population data to detect hidden hotspots.

Invest in Climate Resilience

* Promote water harvesting, small-scale irrigation, and soil conservation to reduce drought risks.

* Introduce community-level crop insurance schemes against drought and pests.
